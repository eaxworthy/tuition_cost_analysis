```python
from bs4 import BeautifulSoup as bs
import requests
import re

url = 'https://nces.ed.gov/fastfacts/display.asp?id=76'
response = requests.get(url)

#html = response.read().decode('cp1252')
htmlsoup = bs(response.content, "html.parser")
table = htmlsoup.find_all('tr')

rows = [] 
for x in htmlsoup.find_all('table'):
    rows.extend(
         (i for i in x.find_all('tr'))
    )
rows
```

```python
text = []
for item in rows:
    s = item.get_text()
    s = s.strip()
    s = s.replace(',','')
    text.append(s)
text
```

```python
all_institutions = [item.split() for item in text[4:25]]
all_institutions
```

```python
public_institutions = [item.split() for item in text[26:46]]
public_institutions
```

```python
private_institutions = [item.split() for item in text[47:]]
private_institutions
```

```python
#We then trim down the incorrect lines for private institutions with another list comprehension
private_institutions = [item[0:7] for item in private_institutions]
private_institutions
```

```python
#combine the three lists into one so we can save as a csv
#the second tier of table headers weren't grabbed by beautiful soup, so we're adding them manually. There are \\
#only six, so it's a better use of time to do it this way rather than try to write a script.
headers = [['Category','Year', 'Constant_All', 'Constant_4-Year', 'Constant_2-year', 'Contemporary_All', 'Contemporary_4-Year', 'Contemporary_2-Year']]
if all_institutions[0][0] != 'all':
    all_institutions = [['All Institutions']+item for item in all_institutions]
    public_institutions = [['Public']+item for item in public_institutions]
    private_institutions = [['Private']+item for item in private_institutions]
combined = headers+all_institutions + public_institutions + private_institutions
```

```python
combined = [','.join(row) for row in combined]
combined[:5]
```

```python
with open("data/table.csv", 'w', encoding='utf8') as file:
    for row in combined:
        row=row.replace('$','')
        file.write(row + "\n")
```
